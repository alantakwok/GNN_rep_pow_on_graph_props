## Finding the number of epochs the net took to converge

```python
def convergence(a):
    flag = 0
    conv = []
    for i in range(1,len(a)):
        if (a[i-1] - a[i]) < 10e-8:
            conv.append(i)
            flag += 1
        if flag >= 300:    
            return(conv[0])
        else:
            conv = []
```

```python
new_AT = [centrality_1111_norm_AT , cluster_coef_1111_norm_AT, ecentricity_1111_norm_AT, neighbor_degree_1111_norm_AT, page_rank_1111_norm_AT, shortest_path_length_1111_norm_AT]
new = [centrality_1111_norm , cluster_coef_1111_norm, ecentricity_1111_norm, neighbor_degree_1111_norm, page_rank_1111_norm, shortest_path_length_1111_norm]

at = []
nat = []
for i in range(len(new_AT)):
    at.append(convergence(new_AT[i]['history']['val_loss']))
    nat.append(convergence(new[i]['history']['val_loss']))
```
